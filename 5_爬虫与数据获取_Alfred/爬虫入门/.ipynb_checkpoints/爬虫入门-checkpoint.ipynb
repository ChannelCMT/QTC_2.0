## 2. lxml

### 新浪财经个股上龙虎榜统计的网站的案例

#### 传递 URL 参数，生成名为 r 的 Response 对象。我们可以从这个对象中获取所有我们想要的信息

```python
# 发送请求，获取响应
r = requests.get(url)
```

```python
r
```

```python
r.content
```

```python
## read到这步跟content是一样的
```

```python
from lxml import etree
```

```python
print(r.content.decode("GBK"))
```

```python
html = etree.HTML(r.content)
```

```python
head = html.xpath("//table[@id=\"dataTable\"]/thead/tr")
```

```python
head
```

```python
[child.text for child in head[0].getchildren()]
```

```python
head[0].xpath('string(.)')  # 拿到所有的文本
```

```python
print(head[0].xpath('string(.)'))
```

```python
import re
re.split(r"[~\r\n\t ]+", head[0].xpath('string(.)'))
```

```python
columns = [col for col in re.split(r"[~\r\n\t ]+", head[0].xpath('string(.)')) if bool(col)]
```

```python
columns
```

#### 把之前那个方法写下来

```python
res = html.xpath("//table[@id=\"dataTable\"]/tr")
```

```python
res
```

```python
def get_str_list(element):
    texts = element.xpath('string(.)')
    text_list = re.split(r"[~\r\n\t ]+", texts)
    return [text for text in text_list if bool(text)]
```

```python
get_str_list(res[0])
```

### 转换成数据

```python
import pandas as pd
pd.DataFrame(
    [get_str_list(_res) for _res in res],
    columns = columns
)
```

### 另一种做法

```python
sarr = [etree.tostring(node).decode('utf-8') for node in res]
```

```python
sarr = ''.join(sarr)
```

```python
sarr = '<table>%s</table>'%sarr
```

```python
sarr
```

```python
df = pd.read_html(sarr)[0]
```

```python
df
```

## 200 403
需要添加HTTP头部，伪装成浏览器

#### 只要简单地传递一个 dict 给 headers 参数就可以了。

```python
agent = 'Mozilla/5.0 (Windows NT 5.1; rv:33.0) Gecko/20100101 Firefox/33.0'
headers = {'Host': 'xueqiu.com',
           'Referer': 'https://xueqiu.com/',
           'Origin':'https://xueqiu.com',
           'User-Agent': agent}
```

```python
xueqiu = requests.get("https://xueqiu.com/hq/screener", headers=headers)
```

```python
etree.HTML(xueqiu.content).xpath('//*[@id="stock-rank"]/div[3]/table/tbody/tr[1]/td[1]/a')
```

```python
xueqiu.encoding
```
